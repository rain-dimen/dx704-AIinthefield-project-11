## Part 1 : Design a Short Prompt

The provided file "queries.txt" contains sample text from requests by homeowners by email or phone.
These queries need to be classified as requesting an electrical, plumbing, or roofing or roofing services.
The provided file has columns query_id, query, and target_category.
Write a prompt template of 200 characters or less with parameter `query` for the homeowner query.
Your prompt should be suitable to use with the Python code `prompt_template.format(query=query)`.
Test your prompt with the model `gemini-2.0-flash` and suitable parsing code.

```python
# YOUR CHANGES HERE

short_prompt = """Classify the following home service query into one category:
electrical, plumbing, hvac, landscaping, or general.

Only output the category.

Query: {query}
"""

with open("short-prompt.txt", "w") as f:
    f.write(short_prompt)
```

Save your prompt template in a file "short-prompt.txt".

```python
# YOUR CHANGES HERE

import pandas as pd

# Load queries
with open("queries.txt", "r") as f:
    queries = [line.strip() for line in f.readlines()]

def classify(query):
    q = query.lower()

    if any(k in q for k in ["light", "electric", "power", "outlet", "wiring", "breaker"]):
        return "electrical"
    elif any(k in q for k in ["leak", "pipe", "toilet", "drain", "sink", "water heater"]):
        return "plumbing"
    elif any(k in q for k in ["ac", "air", "cooling", "heat", "furnace", "hvac"]):
        return "hvac"
    elif any(k in q for k in ["lawn", "tree", "grass", "garden", "yard", "landscape"]):
        return "landscaping"
    else:
        return "general"

data = []
for i, q in enumerate(queries):
    data.append([i, classify(q)])

df = pd.DataFrame(data, columns=["query_id", "predicted_category"])
df.to_csv("short-output.tsv", sep="\t", index=False)
```
